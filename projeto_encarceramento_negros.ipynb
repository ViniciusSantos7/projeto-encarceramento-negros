```python
import pandas as pd

df = pd.read_csv('encarceramento_rj_sintetico.csv')
df.head()
```

```python
df.info()
```

```python
df["Proporcao_Presos_Negros"] = df["Presos_Negros"] / df["Presos_Totais"]

df[["Municipio", "Presos_Totais", "Presos_Negros", "Proporcao_Presos_Negros"]]
```

```python
# Criar coluna com a proporção da população negra no município
df["Percentual_Negra"] = df["Percentual_Negra"]  # já existe no dataframe, só confirmando

# Criar coluna de diferença entre proporção de presos negros e proporção da população negra
df["Diferenca"] = df["Proporcao_Presos_Negros"] - df["Percentual_Negra"]

# Mostrar tabela resumida com essas informações
df[["Municipio", "Percentual_Negra", "Proporcao_Presos_Negros", "Diferenca"]]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="Municipio", y="Percentual_Negra", data=df, color="lightblue", label="População Negra")
sns.barplot(x="Municipio", y="Proporcao_Presos_Negros", data=df, color="red", alpha=0.6, label="Presos Negros")
plt.legend()
plt.title("Comparação entre população negra e presos negros por município")
plt.ylabel("Proporção")
plt.xlabel("Município")
plt.ylim(0, 1)
plt.show()
```
